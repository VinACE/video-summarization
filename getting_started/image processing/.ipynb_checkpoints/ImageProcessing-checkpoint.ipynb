# Reading and Displaying Image

```python
import cv2  # OpenCV
from matplotlib import pyplot as plt  # to display images in Jupyter Notebook
```

```python
image = cv2.imread('dolphins.jpg', 1)  # read image
```

```python
# OpenCV by defaults reads image in BGR but matplotlib by default interprets image as RGB.
# So we need to convert the image to RGB for proper rendering
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB);
```

```python
plt.imshow(image)  # display image
```

# Getting Shape and Size of an Image

```python
height_image1, width_image1, channels_image1 = image.shape  # getting shape of an image
print(height_image1, width_image1, channels_image1)
```

# Cropped Image

```python
plt.imshow(image[100:250, 30:300]) # display cropped image
```

# Blending Two Images (Adding Two Images)

```python
image2 = cv2.imread("lion.jpg")  # read the second image
```

```python
image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2RGB)  # convert it to RGB
```

```python
# Create blended image according to your taste
blendedImage = image[:] * 0.75 + image2[100:100+height_image1, 250:250+width_image1] * 0.25
```

```python
plt.imshow(blendedImage.astype(int))  # show the blended image
```
